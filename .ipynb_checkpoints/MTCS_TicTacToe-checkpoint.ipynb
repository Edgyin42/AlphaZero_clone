```python
PLAYER_X = 'X'
PLAYER_O = 'O'
EMPTY_SPOT = '_'

class TicTacToe: 
    def __init__(self): 
        self.current_player = PLAYER_X
        self.board = np.array([[EMPTY_SPOT]*3 for _ in range(3)])
        self.winner = None 
    def play(self, i, j): 
        '''Choose the row and col of the move
        i: row
        j: col
        '''
        if self.board[i][j] != "_":  
            print("This move is played already. Choose another spot") 
            return 
        if self.turn: 
            self.board[i][j] = "X" 
        else: 
            self.board[i][j] = "O" 
        self.checkWin()
        if self.winPlayer is not None: 
            if self.winPlayer: 
                print("X wins the game") 
            else: 
                print("O wins the game") 
        self.turn = not self.turn
        
    def checkWin(self): 
        '''Check whether either player has won the game'''
        
        for i in range(3): 
            if self.board[i][0] == self.board[i][1] == self.board[i][2] and self.board[i][0] != "_": 
                self.winPlayer = True if self.board[i][0] == "X" else False
            if self.board[0][i] == self.board[1][i] == self.board[2][i] and self.board[0][i] != "_": 
                self.winPlayer = True if self.board[0][i] == "X" else False
            if self.board[0][0] == self.board[1][1] == self.board[2][2] and self.board[0][0] != "_": 
                self.winPlayer = True if self.board[0][0] == "X" else False
            if self.board[0][2] == self.board[1][1] == self.board[2][0] and self.board[0][2] != "_": 
                self.winPlayer = True if self.board[0][2] == "X" else False      
                
    def find_possible_moves(self): 
        
        possible_moves = []
        for i in range(3): 
            for j in range(3):
                if self.board[i][j] == "_":
                    possible_moves.append((i, j))
        return possible_moves
```
